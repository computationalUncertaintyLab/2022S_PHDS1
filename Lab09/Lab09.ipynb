### The Law of large numbers  and the CLT

We can experiment with the theory of the law of large numbers and the more powerful central limit theorem by simulating random variables in R.   
First we will see how the law of large numbers (LLN) can be used to approximate, to arbitrary percision, probabilities, the expectiation, and the variance. 
Next, we will show that thge sample mean of many different types of random variables all have a normal distribution. 

The LLN states for a sequence of independent identically distributed random variables $X_{1},X_{2},X_{3}, \cdots, X_{N}$, the new random variable $\overline{X_{n}}$ will approach $\mathbb{E}(X)$. 

Suppose $X_{1}, X_{2}, \cdots, X_{N} \sim \text{Pois}(\lambda)$.
The LLN then states that as $N$ increases the sample mean $\overline{X_{n}} = \frac{X_{1} + X_{2} + \cdots + X_{N}}{N}$ approaches the expected value $\mathbb{E}(X) = \lambda$. 

But what do we mean by approach? 

The $\overline{X_{n}}$ is a **random variable** and so has a range of possible values and probabilties assigned to each value. 
To be more precise about what we mean when we say "approach", the LLN states that as N increases the *probability* $P \left(| \overline{X_{n}} - \mathbb{E}(\overline{X}) | \right) \to 0$

Lets test the LLN. 
We will simulate 5, 6, 7, $\cdots$ 5,000 random variables with a Poisson distribution with parameter 2.0 and compute 

```r
simulate_average_values_for_pois = function(parameter,N,M=1000){
    avg_value = c()
    for (n in 1:N){
        samples = rpois(M,parameter)
        mean_sample = mean(samples)
        avg_value = c(avg_value,mean_sample)
    }
    return(avg_value)
}

fifty_samples = simulate_average_values_for_pois(2,50)
five_hundred_samples   = simulate_average_values_for_pois(2,500)
five_thousand_samples  = simulate_average_values_for_pois(2,5000)

par(mfrow=c(1,3))

hist(fifty_samples        , 20, xlab = "X 50")
hist(five_hundred_samples , 20, xlab = "X 500")
hist(five_thousand_samples, 20, xlab = "X 5000")
```

We see that as $N$ increases this probability distribution becomes more peaked over the parameter value $2$. 
Another way to visulalize the LLN is to perform the following:
1. For values of N from 1 to (say) 300:
    2. Generate 100 realizations of $\overline{X_{n}}$
    3. Plot the 100 values (N,$\overline{X_{n}}$)


```r
plot(1:300,rep(2,300), type="l", lwd=1, xlab="N", ylab="Xbar")
mycol <- rgb(0, 0, 255, max = 255, alpha = 25, names = "blue50")

for (num in 1:200){
    avgs = c()
    for (n in 1:300){
       samples = rpois(n,2)
       avg     = mean(samples)
       points(n,avg, col= mycol, cex=0.75 )    
    }
}
```

As expected by the LLN, as N increase our samples of $\overline{X}$ become more and more concentratrated around the expected value of $\overline{X}$. Because $X_{i} \sim \text{Pois}(2)$ the LLN states that these realizations should become more and more concentrated around 2.
