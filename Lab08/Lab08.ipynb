### Simulating realizations from random variables.

Random variables can be difficult to imagine in practice. 
However, the computer allows us to draw values from a random variable according to a specific probability distribution to gain understanding about these distributions and to make random variables a bit more tangible.

#### Simulating the Binomial Distribution

A random variable $Y$ has a binomial distribution with parameters $N$ and $\theta$. 
With R we can generate a single realization, or data point, from the random variable $Y$ by executing the ``rbinom`` function. 

The ``rbinom`` function takes as input 3 arguments. The first argument is the number of data points to generate, the second input is the parameter $N$, and the third argument is the parameter $\theta$.

We can generate a single data point for the random variable $Y \sim \text{Binom}(10,0.2)$ like this

```r
datapoint = rbinom(1, 10, 0.2)
print(datapoint)
```

We can generate 2000 data points from 2000 random variables with the same exact distribution $Y_{1}, Y_{2}, \cdots, Y_{2000} \sim \text{Binom}(10, 0.2)$ like this

```r
datapoints = rbinom(2000, 10, 0.2)
```

R will generate 2000 realizations and store this as a vector. 

```r
print(datapoints)
```

We can visualize the frequency of these datapoints with a histogram. 

```r
hist(datapoints)
```

#### Simulating the Bernoulli Distribution

A bernoulli random variable $X \sim \text{Bern}(\theta)$ is the same as a random variable that has a binomial distribution with the parameter $N$ = 1. 

Suppose $Y \sim \text{Bin}(1,\theta)$. 
The $supp(Y) = \{0,1\}$ and the probability mass function is 

\begin{align}
    f_{Y}(y) = \binom{1}{y} \theta^{y} (1-\theta)^{1-y}
\end{align}

When $y = 0$

\begin{align}
    f_{Y}(0) &= \binom{1}{0} \theta^{0} (1-\theta)^{1-0}\\
    f_{Y}(0) &= 1 \times 1 (1-\theta)^{1}\\
    f_{Y}(0) &= (1-\theta)\\
\end{align}

and when $y=1$

\begin{align}
    f_{Y}(1) &= \binom{1}{1} \theta^{1} (1-\theta)^{1-1}\\
    f_{Y}(1) &= 1 \times \theta (1-\theta)^{0}\\
    f_{Y}(1) &= \theta\\
\end{align}

The probability mass function for $Y \sim \text{Binom}(1,\theta)$ is equal to the pmf for a bernoulli distribution, and so $Y \sim \text{Bern}(\theta)$.  

We can simulate in R 2000 random variables $Y_{1},Y_{2},Y_{3}, \cdots, Y_{2000} \sim \text{Bern}(0.4)$ using the ``rbinom`` function with parameter $N=1$.

```r
datapoints = rbinom(2000, 1, 0.4)
hist(datapoints)
```

#### Simulating the Poisson Distribution

We can simulate data from a set of poisson random variables using the ``rpois`` function.
The ``rpois`` function has two arguments. The first argument is the number of data points to generate and the second data point is the parameter associated with the poisson distribition.

If we wanted to generate 100 random variables $S_{1},S_{2}, \cdots, S_{100} \sim \text{Pois}(2.2)$ we can write

```r
datapoints = rpois(100, 2.2)
hist(datapoints)
```

#### Simulating the Normal Distribution

The ``rnorm`` function is used to simulate data from a set of Normal random variables. 
The ``rnorm`` function takes three arguments. The first argument is the number of datapoints to generate, the second argument is the parameter $\mu$ of the normal distribution and the third argument is the parameter $\sigma$ (not $\sigma^{2}$) of the normal distribution. 

Below we use the ``rnorm`` function to generate 2000 data points from a set of 2000 normally distributed random variables with $\mu=2$ and $\sigma=1$.

```r
samples = rnorm(2000, 2, 1)
hist(samples)
```

## Transformation

Often the exact probability mass (density) function for a transformation of random variables can be difficult to write down, or even impossible to write down.
We can estimate transformations of random variables by (i) simulating $N$ datapoints, (ii) transforming each sample, and (iii) storing these transformed data points as an approximate probability density.  


For example, suppose we want to understand the distribution of $Y = X^2$ where $X \sim \mathcal{N}(0,1)$.
(Note that this distribution does have a pdf you can write down). 

Lets work through the steps of approximating the pdf of $Y$.
Step 1: Generate $N$ samples from $X$. (lets set N to 5000)

```r
Xsamples = rnorm(5000,0,1)
hist(Xsamples)
```

Step 2: Transform each of the ``Xsamples``

```r
Y = Xsamples^2
```

Step 3: Visualize the pdf

```r
hist(Y)
```

### The empirical CDF

The cumulative density function (cdf) is important for computing the probability a random variable $X$ assigns to values less than $x$ for $x \in supp(X)$.
Like the histogram approximates the probability density (or mass) function of a random variable, we can approximate the cumulative density function too. 


Given a set of data points $(x_{1}, x_{2}, \cdots, x_{N} )$ generated from $N$ identical copies of a random variable with probability density $f_{X}$,  $X_{1}, X_{2}, \cdots, X_{N} \sim f$, the **empirical cumulative density function** evaluated at the value $x$---$F_{X}(x)$---is computed as  

\begin{align}
    F_{X}(x) = \frac{\sum_{i=1}^{N} \mathbb{1}( x_{i} < x )}{N}
\end{align}

The function $\mathbb{1}(x)$ is called an indicator function. An indicator function returns the value $1$ when the input (x) is true and the value $0$ when the input (x) is false. 
The empirical cdf function~(sometimes abbreviated the ecdf) assigns to $x$ the proportion of data points that are smaller than $x$. 

Lets compute the the ecdf for the above random variable $Y$. 

Step 1: Pick a sequence of values for the ECDF. Lets choose 100 points between the minimum simulated value and maximum simulated value. The ``seq`` function is a handy tool for picking equally spaced points between two values. 

```r
min_value = min(Y)
max_value = max(Y)

values = seq(min_value, max_value, (max_value-min_value)/100)
```

Step 2: Iterate a variable $x$ through all ``values`` and compute the proportion of datapoints that are smaller than $x$. Store them in a vector called ``ecdf``

```r
ecdf = c()

for (x in values){
    prob = mean(Y<x) # proportion below x
    ecdf = c(ecdf, prob)
}
```

Step 3: Plot the ecdf!

```r
library(ggplot2)

d = data.frame(x = values, y = ecdf)

ggplot(data = d, aes(x = x, y = y)) + 
geom_point() +
geom_line(linetype = 'dashed')

#plot(values, ecdf, xlab = "y", ylab = "P(Y <y)")
#lines(values, ecdf)

```

#### Inverse CDF sampling: one method to simulate a random variable

We have seen how functions in R can generate realizations or draws from a random variable with minimal code, but we have not yet explored **how** R does this.
One method to generate realizations from a random variable is called the inverse CDF method. 

The inverse CDF method relies on the ability to generate values uniformly at random between the interval bounded below by zero and above by one and that for the distribution we wish to draw samples from that we can compute the inverse CDF. 

##### The exponential distribution 

A continuous random variable has an exponential distribution with parameter $\lambda$ if the support of $X$ is all positive real numbers $supp(X) = \mathbb{R}^{+}$ and the probability density function is 

\begin{align}
       f_{X}(x) = \lambda e^{-\lambda x}
\end{align}

The expected value and variance of the exponential distribution is

\begin{align}
    \mathbb{E}(X) &= \frac{1}{\lambda} \\ 
             V(X) &= \frac{1}{\lambda^{2}} \\ 
\end{align}

The cumulative density function for the exponential distribution is 

\begin{align}
       F_{X}(x) = P(X<x) = 1 - e^{-\lambda x}
\end{align}

To sample from the exponential distribution we can do the following:
1. Compute the inverse CDF (Q)---a function that maps values from [0-1] to values the exponential distribution generates
2. Generate a random uniform value $u$ between 0 and 1. 
3. Consider the value Q(u) a randomly generated value from our exponential distribution.
